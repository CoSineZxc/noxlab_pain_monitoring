```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_pickle('../Data/descriptive_data/DescriptiveData.pkl')
```

```python
data
```

```python
fano_threshold = []
for i in range(len(data['Fano factor'])):
    fano_threshold.append(np.array(data['Fano factor'][i]) > 0.2)
```

```python
retained_ratings = []
retained_timings = []
for i in range(len(data)):
    retained_ratings.append(np.array(data['ratings'].iloc[i])[fano_threshold[i]])
    retained_timings.append(np.array(data['time'].iloc[i])[fano_threshold[i]])
```

```python
participant_list = data['PID']
prediction_all = data['Prediction']
confidence_all = data['Confidence']
```

```python
data_dict = {'PID': participant_list, 'ratings': retained_ratings, 'time': retained_timings, 'Prediction': prediction_all, 'Confidence': confidence_all}
data_descriptive = pd.DataFrame(data_dict)
data_descriptive.to_pickle('../Data/data_file/FilteredLowVarianceData.pkl')
```
