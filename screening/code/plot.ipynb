### plot distribution

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
df=pd.read_csv(r"../data/qualtrics_05_excludeother_score.csv")
```

```python
# plot score distribution
df['Score'].plot(kind="hist",bins=10,color="steelblue",edgecolor="black")
x_major_locator=MultipleLocator(5)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
plt.xlim(5,60)
plt.xlabel("score")
plt.title("Score distribution")
plt.show()
```

```python
# add kernel density estimation
df['Score'].plot(kind="hist",bins=10,color="steelblue",edgecolor="black",density=True)
df['Score'].plot(kind="kde",color="red")
plt.xlabel("score")
plt.title("Score distribution")
x_major_locator=MultipleLocator(5)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
plt.xlim(5,60)
plt.show()
```
