```python
import pandas as pd
import nolds
import numpy as np
import math
from tqdm.notebook import tqdm
import ptitprince as pt
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ttest_ind
```

```python
def datadiv(data,div):
    data_np=np.array(data)/div
    data=data_np.tolist()
    return data
```

```python
pain_rating_df=pd.read_pickle('../data/CBP_rating.pkl')
Carl_all_rating=[]
day_list=[]
for index,row in pain_rating_df.iterrows():
    if row.D1_rating!=[]:
        Carl_all_rating.append(datadiv(row.D1_rating,10))
        day_list.append(1)
    if row.D2_rating!=[]:
        Carl_all_rating.append(datadiv(row.D2_rating,10))
        day_list.append(2)
    if row.D3_rating!=[]:
        Carl_all_rating.append(datadiv(row.D3_rating,10))
        day_list.append(3)
    if row.D4_rating!=[]:
        Carl_all_rating.append(datadiv(row.D4_rating,10))
        day_list.append(4)
Sharonallrating_df=pd.read_pickle('../data/Sharon_allrating.pkl')
Sharon_all_rating=Sharonallrating_df['Sharon rating'].tolist()
```

```python
mean_Carl=[]
for rating in Carl_all_rating:
    mean_Carl.append(np.mean(rating))
mean_Sharon=[]
for rating in Sharon_all_rating:
    mean_Sharon.append(np.mean(rating))
plt.xlabel("mean rating")
plt.title("Carl's mean rating")
plt.hist(mean_Carl, bins=20)
plt.xlim(0,10)
plt.show()
plt.xlabel("mean rating")
plt.title("Sharon's mean rating")
plt.hist(mean_Sharon, bins=20)
plt.xlim(0,10)
plt.show()
mean_rating_carl_np=np.array(mean_Carl)
mean_rating_sharon_np=np.array(mean_Sharon)
ttest_ind(mean_rating_carl_np,mean_rating_sharon_np)
```

```python
std_Carl=[]
for rating in Carl_all_rating:
    std_Carl.append(np.std(rating))
std_Sharon=[]
for rating in Sharon_all_rating:
    std_Sharon.append(np.std(rating))
plt.xlabel("std")
plt.title("Carl's standard deviation")
plt.hist(std_Carl, bins=50)
plt.xlim(0,3.5)
plt.show()
plt.xlabel("std")
plt.title("Sharon's standard deviation")
plt.hist(std_Sharon, bins=50)
plt.xlim(0,3.5)
plt.show()
std_rating_carl_np=np.array(std_Carl)
std_rating_sharon_np=np.array(std_Sharon)
ttest_ind(std_rating_carl_np,std_rating_sharon_np)
```

```python
def cv(data):
    mean=np.mean(data)
    std=np.std(data)
    cv=std/mean
    return cv
```

```python
cv_Carl=[]
num=0
for rating in Carl_all_rating:
    cv_Carl.append(cv(rating))
    if cv(rating)>2.5:
        num+=1
print(num)
cv_Sharon=[]
for rating in Sharon_all_rating:
    cv_Sharon.append(cv(rating))
plt.xlabel("cv")
plt.title("Carl's coefficient of variation")
plt.hist(cv_Carl, bins=300)
# plt.xlim(0,2.5)
plt.show()
plt.xlabel("cv")
plt.title("Carl's coefficient of variation")
plt.hist(cv_Carl, bins=300)
plt.xlim(0,2.5)
plt.show()
plt.xlabel("cv")
plt.title("Sharon's coefficient of variation")
plt.hist(cv_Sharon, bins=50)
plt.xlim(0,2.5)
plt.show()
std_rating_carl_np=np.array(std_Carl)
std_rating_sharon_np=np.array(std_Sharon)
ttest_ind(std_rating_carl_np,std_rating_sharon_np)
```
