# PCS

```python
data_pcs = data_1_filtered[['PCS_1',
       'PCS_2', 'PCS_3', 'PCS_4', 'PCS_5', 'PCS_6', 'PCS_7', 'PCS_8', 'PCS_9',
       'PCS_10', 'PCS_11', 'PCS_12', 'PCS_13']]
data_pcs
```

```python
data_pcs = data_pcs.replace('0 - not at all', 0)
data_pcs = data_pcs.replace('1 - to a slight degree', 1)
data_pcs = data_pcs.replace('2 - to a moderate degree', 2)
data_pcs = data_pcs.replace('3 - to a great degree', 3)
data_pcs = data_pcs.replace('4 - all the time', 4)
```

```python
total_pcs_score = []
nan_index = []
for i in range(len(data_1_filtered)):
    if data_1_filtered[['PCS_1',
       'PCS_2', 'PCS_3', 'PCS_4', 'PCS_5', 'PCS_6', 'PCS_7', 'PCS_8', 'PCS_9',
       'PCS_10', 'PCS_11', 'PCS_12', 'PCS_13']].iloc[i].isna().any():
            nan_index.append(i)
            total_pcs_score.append(np.nan)
    else:
          total_pcs_score.append(np.sum(data_pcs[['PCS_1',
       'PCS_2', 'PCS_3', 'PCS_4', 'PCS_5', 'PCS_6', 'PCS_7', 'PCS_8', 'PCS_9',
       'PCS_10', 'PCS_11', 'PCS_12', 'PCS_13']].iloc[i]))
```

```python
rumination_pcs_score = []
nan_index = []
for i in range(len(data_1_filtered)):
    if data_1_filtered[['PCS_8', 'PCS_9',
       'PCS_10', 'PCS_11']].iloc[i].isna().any():
            nan_index.append(i)
            rumination_pcs_score.append(np.nan)
    else:
          rumination_pcs_score.append(np.sum(data_pcs[['PCS_8', 'PCS_9',
       'PCS_10', 'PCS_11']].iloc[i]))
```

```python
magnification_pcs_score = []
nan_index = []
for i in range(len(data_1_filtered)):
    if data_1_filtered[['PCS_6', 'PCS_7', 'PCS_13']].iloc[i].isna().any():
            nan_index.append(i)
            magnification_pcs_score.append(np.nan)
    else:
          magnification_pcs_score.append(np.sum(data_pcs[['PCS_6', 'PCS_7', 'PCS_13']].iloc[i]))
```

```python
data_1_filtered[['PCS_6', 'PCS_7', 'PCS_13']].iloc[nan_index]
```

```python
helplessness_pcs_score = []
nan_index = []
for i in range(len(data_1_filtered)):
    if data_1_filtered[['PCS_1',
       'PCS_2', 'PCS_3', 'PCS_4', 'PCS_5', 'PCS_12']].iloc[i].isna().any():
            nan_index.append(i)
            helplessness_pcs_score.append(np.nan)
    else:
          helplessness_pcs_score.append(np.sum(data_pcs[['PCS_1',
       'PCS_2', 'PCS_3', 'PCS_4', 'PCS_5', 'PCS_12']].iloc[i]))
```

```python
data_1_filtered['Total PCS Score'] = total_pcs_score
data_1_filtered['PCS Rumination Score'] = rumination_pcs_score
data_1_filtered['PCS Magnification Score'] = magnification_pcs_score
data_1_filtered['PCS Helplessness Score'] = helplessness_pcs_score
data_1_filtered
```
